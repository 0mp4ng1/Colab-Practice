## 단순 w를 0~40 노가다로 구하기
forward 함수 : X --> MODEL --> Y 

```python
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0, 2.0, 3.0]
y_data = [20.0, 40.0, 60.0]

def forward(x):
  return w * x #y_hat

def loss(x, y): #Squared Error
  y_pred = forward(x)
  return (y_pred - y) ** 2

# w를 0부터 40까지 노가다를 해보자!
for w in np.arange(0.0, 40.0, 1.0):
  l_sum = 0
  for x, y in zip(x_data, y_data):
    l_sum += loss(x, y)
  l_sum /= 3
  # l_sum is MSE (평균)
  print(f'w: {w}, loss: {l_sum}')
```

## Gradient Descent

```python
# Gradient Descent
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0, 2.0, 3.0]
y_data = [20.0, 40.0, 60.0]

def forward(x):
  return w * x #y_hat

def loss(x, y): #Squared Error
  y_pred = forward(x)
  return (y_pred - y) ** 2

w = 0
lr = 0.01 # learning rate

# loss = (w * x - y) ^ 2
# dloss / dw = 2 * x * (w * x - y)
def gradient (x, y):
  return 2 * x * (w * x - y)

# 몇 "번"? = 100 epoch (학습 횟수)
for epoch in range(100):
  l = 0
  grad = 0
  for x, y in zip(x_data, y_data):
    grad += gradient(x, y)
    grad /= 3
    l += loss(x, y)
  w = w - lr * grad
  l /= 3

  print(f'epoch : {epoch}, w : {w}, l : {l}')
```

### Back Propagation 사용

```python
import torch

x_data = [1.0, 2.0, 3.0]
y_data = [20.0, 40.0, 60.0]

w = torch.tensor([1.0], requires_grad = True)

# our model forward
def forward(x):
  return w * x #y_hat

def loss(x, y): #Squared Error
  y_pred = forward(x)
  return (y_pred - y) ** 2

lr = 0.01
# 몇 "번"? = 100 epoch (학습 횟수)
for epoch in range(100):
  l = 0
  for x, y in zip(x_data, y_data):
    l += loss(x, y)

  l.backward()
  w.data = w.data - lr * w.grad.item()

  w.grad.data.zero_() # w.grad.itme() -> 0
  
  print(f'epoch : {epoch}, w : {w.data}')
```

## MODEL 이용 

```python
import torch
import torch.nn as nn
import torch.optim as optim

x_data = torch.Tensor([1.0, 2.0, 3.0])
y_data = torch.Tensor([20.0, 40.0, 60.0])

class Model(nn.Module):
  def __init__(self):
    super().__init__()
    self.w = torch.tensor([1.0], requires_grad = True)

  def forward(self, x):
    return self.w * x

loss = nn.MSELoss()
model = Model()

lr = 0.01

for epoch in range(100):
  y_pred = model(x_data) # model.forward(x_data)
  error = loss(y_pred, y_data)
  error.backward()
  model.w.data = model.w.data - lr * model.w.grad.item()

  model.w.grad.data.zero_() # w.grad.itme() -> 0

  print(f'epoch : {epoch}, w : {model.w.data}')
```
